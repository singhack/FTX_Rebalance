```python
time_period =100 
returns_multiples =[]
corr_multiples =[]



symbol_lists= ['1INCH/USD','SOL/USD','SRM/USD','XRP/USD','ETH/USD','AAVE/USD','UNI/USD','ALPHA/USD','BNB/USD','FTT/USD','LINA/USD','DOGE/USD','BAL/USD','MATIC/USD']
```

```python
timeframes ='4h'
start_date = datetime.datetime.now() - datetime.timedelta(days=100)
start_ts = start_date.timestamp()*1000
end_date = datetime.date.today()

index_name = 'BTC/USD'
index_df = pd.DataFrame(ex.fetch_ohlcv(index_name,timeframe=timeframes, since=start_ts),columns=['date','open','high','low','close','volume'])
index_df['Percent Change'] = index_df['close'].pct_change()
index_df['date'] = pd.to_datetime(index_df['date'],unit='ms')
index_df.set_index = index_df.set_index('date',inplace=True)

index_return = (index_df['Percent Change'] + 1).cumprod().iloc[-1]

try:
    for symbol in tqdm(symbol_lists):
        # Download historical data as CSV for each stock (makes the process faster)
        df = pd.DataFrame(ex.fetch_ohlcv(symbol,timeframe=timeframes, since=start_ts),columns=['date','open','high','low','close','volume'])
    # Calculating returns relative to the market (returns multiple)
        df['date'] = pd.to_datetime(df['date'],unit='ms')
        df.set_index = df.set_index('date',inplace=True)

        df['Percent Change'] = np.log(df['close']/df['close'].shift(1)).fillna(0)
        stock_return = (df['Percent Change'] + 1).cumprod().iloc[-1]
        returns_multiple = round((stock_return / index_return), 2)
        returns_multiples.extend([returns_multiple])
        corr = df['Percent Change'].corr(index_df['Percent Change'] )
        corr_multiples.extend([corr])
        print (f'Symbol: {symbol}; Periods {time_period} Days  Returns Multiple against {index_name}: {returns_multiple} Correlation:{corr:.2f}\n')
#         df.to_csv(f'{symbol}.csv')

#         time.sleep(0.5)
except Exception as e:
    print(str(e))
    

```

```python
rs_df = pd.DataFrame(list(zip(symbol_lists, returns_multiples)), columns=['Symbols', 'Returns_multiple'])
rs_df['RS_Rating'] = rs_df.Returns_multiple.rank(pct=True) * 100
rs_df = rs_df[rs_df.RS_Rating >= rs_df.RS_Rating.quantile(.5)]  #### Filter abover 90 percentile
corr_df = pd.DataFrame(list(zip(symbol_lists, corr_multiples)), columns=['Symbols', 'Corr_multiple'])
corr_df['Corr_Rating'] = corr_df['Corr_multiple'].rank(pct=True)*100
corr_df=corr_df[corr_df.Corr_Rating >= corr_df.Corr_Rating.quantile(.05)]
rs_df
```

```python
#### Screening conditions 
exportList = pd.DataFrame(columns=['Symbols','Start', "RS_Rating",'rsi', "SMA_10", "SMA_20", "SMA_50", "52 Week Low", "52 week High"])

rs_tokens = rs_df['Symbols']
for symbol in rs_tokens:    
    try:
#         df = pd.read_csv(f'{stock}.csv', index_col=0)
        df = pd.DataFrame(ex.fetch_ohlcv(symbol,timeframe='1d', since=start_ts),columns=['date','open','high','low','close','volume'])
        df['date'] = pd.to_datetime(df['date'],unit='ms')
        df.set_index = df.set_index('date',inplace=True)
        df['Percent Change'] = np.log(df['close']/df['close'].shift(1)).fillna(0)
        ret= df['Percent Change'].mean()
        ann_ret = ret *252
        m_ret =ret*30
        corr = df['Percent Change'].corr(index_df['Percent Change'])
        sma = [10, 20, 50]
        for x in sma:
            df["SMA_"+str(x)] = round(df['close'].rolling(window=x).mean(), 2)
#             print(df)
        df['rsi']= ta.RSI(df['close'],timeperiod=10)

        # Storing required values 
        currentClose = df["close"].iloc[-1]
        first_day = df.index[0].strftime('%Y/%m/%d')
        print(first_day)
        moving_average_10 = df["SMA_10"].iloc[-1]
        moving_average_20 = df["SMA_20"].iloc[-1]
        moving_average_50 = df["SMA_50"].iloc[-1]
        low_of_52week = round(min(df["low"][-50:]), 2)
        high_of_52week = round(max(df["high"][-50:]), 2)
        rsi_10 = df['rsi'].iloc[-1]
        RS_Rating = round(rs_df[rs_df['Symbols']==symbol].RS_Rating.tolist()[0])
        
        try:
            moving_average_50_20 = df["SMA_50"][-30]
        except Exception:
            moving_average_50_20 = 0

        # Condition 1: Current Price > 20 SMA and > 50 SMA
        condition_1 = currentClose > moving_average_20 > moving_average_50
        
        # Condition 2: 20 SMA and > 50 SMA
        condition_2 = moving_average_20 > moving_average_50

        # Condition 3: 200 SMA trending up for at least 1 month
        condition_3 = moving_average_50 > moving_average_50_20
        
        # Condition 4: 10 SMA> 20 SMA and 20 SMA> 50 SMA
        condition_4 = moving_average_10 > moving_average_20 > moving_average_50
           
        # Condition 5: Current Price > 10 SMA
        condition_5 = currentClose > moving_average_10
           
        # Condition 6: Current Price is at least 30% above 52 week low
        condition_6 = currentClose >= (1.3*low_of_52week)
           
        # Condition 7: Current Price is within 25% of 52 week high
        condition_7 = currentClose >= (.75*high_of_52week)
        
        ### addtion
        condition_8 = rsi_10 >= 60
        
        ret = round(ret,3)
        m_ret = round(m_ret,3)
        ann_ret = round(ann_ret,3)
        
        # If all conditions above are true, add stock to exportList
#         if(condition_5 and condition_6 and condition_7 and condition_8):
#         if(condition_1 and condition_2 and condition_3 and condition_4 and condition_5 and condition_6 and condition_7):
        exportList = exportList.append({'Symbols': symbol,'Daily_ret':ret,'Monthly_ret':m_ret,'Yearly_ret':ann_ret,'Start':first_day,"corr":corr, "RS_Rating": RS_Rating ,"SMA_10": moving_average_10, "SMA_20": moving_average_20, "SMA_50": moving_average_50, "52 Week Low": low_of_52week, "52 week High": high_of_52week,"rsi":rsi_10}, ignore_index=True)
        print (symbol + " made the Sahing requirements")
    except Exception as e:
        print (e)
        print(f"Could not gather data on {symbol}")

exportList = exportList.sort_values(by='RS_Rating', ascending=False)
print('\n', exportList)
# writer = ExcelWriter("ScreenOutput.xlsx")
# exportList.to_excel(writer, "Sheet1")
# writer.save()
```
